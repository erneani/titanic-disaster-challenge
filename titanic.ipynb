```python
# Pensando no resultado anterior, é interessante termos uma coluna
# que diz se o usuário é do sexo feminino ou não (ao invés
# de uma coluna com textos informando esse dado,
# uma vez que os modelos geralmente não lidam muito bem
# com informações em texto).
# Logo, criaremos uma função que analisa o sexo do passageiro
# e retorna o valor de 1 ou de 0 (representando sim ou não, 
# respectivamente) em uma coluna chamada isFemale.
def is_female(sex):
    # Checando o sexo do passageiro
    if sex == 'female':
        return 1
    
    return 0
```

```python
train_df['isFemale'] = train_df['Sex'].apply(is_female)
test_df['isFemale'] = test_df['Sex'].apply(is_female)
```

```python
# Podemos observar que existem vários valores nulos
# na coluna Age (idade) e na coluna Cabin (cabine).
# Note também que há um valor nulo em Fare (tarifa).
# Como existe uma quantidade enorme de valores nulos
# na coluna Cabin, vamos removê-la.
# PS: As modificações serão realizadas em ambas as
# variáveis para efetuar a limpeza dos dados.
train_df.drop('Cabin', axis=1, inplace=True)
test_df.drop('Cabin', axis=1, inplace=True)
```

```python
# É perceptível que existe uma grande quantidade de
# valores nulos também na coluna Age (idade).
# Porém, como essa coluna é um número, pode ser
# que seja útil para nosso algoritmo de
# Aprendizado de Máquina.
# Portanto, vamos verificar se há uma correlação alta entre
# a idade e a sobrevivência.
train_df.corr()['Survived']
```

```python
# É possível notar que a idade possui uma correlação
# significante com a chance do passageiro ter sobrevivido.
# Logo, vamos buscar uma forma de substituir esses valores
# nulos por valores significantes para nós.
# Então, vamos exibir um gráfico das idades na embarcação
# e sua taxa de sobrevivência.
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Pclass', y='Age', data=train_df)
```

```python
# Com esse gráfico, conseguimos perceber que a média de
# idade entre os tripulantes da primeira classe é
# maior que os da segunda e terceira classe (provavelmente
# porque pessoas mais velhas tiveram mais tempo para acumular riquezas).
# Note que conseguimos observar as médias de idade para cada uma das classes.
# Se quisermos obter as idades sem prejudicar nossos dados,
# podemos substituir cada um dos valores nulos pela média da idade
# da classe a qual o tripulante pertence.
# Observando o gráfico, percebemos que as idades médias das classes são:
# 1ª classe: 37
# 2ª classe: 28
# 3ª classe: 24
```

```python
# Sabendo disso, vamos criar uma função que executa o algoritmo acima.
def replace_age_values(column):
    Age = column[0]
    Pclass = column[1]
    # Primeiro, checamos se a idade é nula
    if pd.isnull(Age):
        # Caso seja nula, vamos checar a classe
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 28
        else:
            return 24
    
    return Age
```

```python
# Aplicaremos essa função à nossa coluna Age(idade)
# utilizando a coluna Pclass para distinguir a classe
# dos tripulantes.
train_df['Age'] = train_df[['Age', 'Pclass']].apply(replace_age_values, axis=1)
test_df['Age'] = test_df[['Age','Pclass']].apply(replace_age_values, axis=1)
```

```python
# Vamos verificar novamente nosso mapa de valores nulos.
sns.heatmap(train_df.isnull())
```

```python
# Perfeito! Nenhum valor nulo em nossos dados de treino!
# Vamos observar agora nossos dados de teste!
sns.heatmap(test_df.isnull())
```

```python
# Conseguimos limpar todos os valores nulos na coluna Age (idade).
# Porém, ainda temos um valor nulo na coluna Fare (tarifa).
# Vamos substituir esse valor na série.
```

```python
# Como temos uma quantidade grande de valores na coluna Fare (tarifa),
# podemos pegar a média das tarifas para substituir esse valor faltante.
fare_mean = test_df['Fare'].mean()
```

```python
# Agora, criaremos uma função que verifica o valor da coluna Fare (tarifa)
# e retorna a média das tarifas caso o valor seja nulo. Se não, retorna 
# a própria tarifa que recebeu como valor.
def fix_nan_fares(fare):
    if pd.isnull(fare):
        return fare_mean
    return fare
```

```python
test_df['Fare'] = test_df['Fare'].apply(fix_nan_fares)
```

```python
# Opa! Aparentemente deu tudo certo!
# Verifiquemos novamente nosso dataset de teste através do 
# mapa de elementos nulos!
sns.heatmap(test_df.isnull())
```

```python
# Agora que nossos dados estão limpos, hora de treinar nosso
# modelo de Aprendizagem de Máquina!
```

```python
# Aprendendo com os dados
# Nesse momento, nós temos diversas opções de modelos para utilizar.
# Utilizarei o modelo de Floresta Randômica por
# acreditar que seja uma boa forma de solucionar o desafio,
# mas um modelo de Regressão Logística, por exemplo, também seria possível
# (assim como outros diversos modelos)
```

```python
# Importaremos nosso modelo de Floresta Randômica.
from sklearn.ensemble import RandomForestClassifier
```

```python
# Após a importação, precisamos criar uma instância do modelo.
# Contudo, note que existe um parâmetro chamado n_estimators.
# Esse parâmetro nos permite definir um número padrão de estimadores
# para melhorar o desempenho do nosso algoritmo.
# Por enquanto, vamos definir como 200.
random_forest_model = RandomForestClassifier(n_estimators=200)
```

```python
# Agora vamos treinar nosso modelo com os dados de treinamento.
# Para isso, precisamos separar os dados entre os atributos
# para se calcular o resultado e o próprio resultado.
# Como queremos descobrir quem sobreviveu ao naufrágio,
# nossa coluna Survived (sobreviveu) será nosso resultado.
y = train_df['Survived']
```

```python
# Nossas outras colunas, no entanto, servirão para avaliar
# o modelo. Sendo assim, excluiremos todas as colunas que
# não possuírem valores numéricos (novamente, lembre-se que
# os modelos geralmente não lidam bem com textos) e a coluna
# de ID, uma vez que não influencia no fato de  tripulante ser
# salvo ou não.
train_df.head(10)
columns_list = ['Name', 'Ticket', 'Embarked', 'PassengerId']
train_df_values = train_df.drop(columns_list, axis=1)
test_df_values = test_df.drop(columns_list, axis=1)
```

```python
# Agora, vamos, de fato, treinar nosso modelo com nossos dados de treino.
# Passaremos todas as colunas (com exceção de Survived, "sobreviveu", 
# para nosso X).
X = train_df_values.drop('Survived', axis=1)
```

```python
# E treinaremos o modelo.
random_forest_model.fit(X, y)
```

```python
# Uma vez treinado, precisamos prever o resultado
# com nossos dados de teste.
predict = random_forest_model.predict(test_df_values)
```

```python
# Agora, já temos as predições do nosso modelo. Simples e poderoso assim.
# Vamos atribuir nossas predições à coluna Survived (sobreviveu)
# em nosso dataset de teste.
test_df['Survived'] = pd.Series(predict)
```
